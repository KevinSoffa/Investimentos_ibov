# Carteira de Ativos
## Análise

```python
import pandas_datareader.data as web
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


carteira = pd.read_excel('Carteira.xlsx')
display(carteira)
```

## Criando os dataframe de Cotações dos ativos da carteira

```python
cotacao_carteira = pd.DataFrame()

for ativo in carteira['Ativos']:
    cotacao_carteira[ativo] = web.DataReader(
        f'{ativo}.SA', 
        data_source= 'yahoo',
        start='2022-01-01',
        end='2022-11-20'
        )['Adj Close']

display(cotacao_carteira)
```

### Filtrando a tabela e preenchendo as tabelas nulas

```python
cotacao_carteira.info()
```

```python
cotacao_carteira = cotacao_carteira.ffill()
cotacao_carteira.info()
```

## Normalizar cada ação

```python
carteira_norm = cotacao_carteira / cotacao_carteira.iloc[0]
carteira_norm.plot(figsize=(15, 5))
plt.legend(loc='upper left')
```

## Puxando cotações do IBOV

```python
cotacao_ibov = web.DataReader(
        '^BVSP', 
        data_source='yahoo', 
        start='2022-01-01',
        end='2022-11-20'
    )

display(cotacao_ibov)
```

## Criando um dataframe da Carteira com as quantidades de ações

```python
valor_investido = pd.DataFrame()

for ativo in carteira['Ativos']:
    valor_investido[ativo] = cotacao_carteira[ativo] * carteira.loc[carteira['Ativos'] == ativo, 'Qtde'].values[0]

display(valor_investido)
```

## Coparação Carteira X IBOV

```python
valor_investido['Total'] = valor_investido.sum(axis=1)

# Normalizando
valor_investido_norm = valor_investido / valor_investido.iloc[0]
cotacao_ibov_norm = cotacao_ibov / cotacao_ibov.iloc[0]

#Carteira de investimentos
valor_investido_norm['Total'].plot(figsize=(15, 5), label='Carteira')


#Cotação IBOV
cotacao_ibov_norm['Adj Close'].plot(figsize=(15, 5), label='IBOV')

plt.legend()
plt.show()
```

```python
retorno_carteira = valor_investido['Total'][-1] / valor_investido['Total'][0] -1
retorno_ibov = cotacao_ibov['Adj Close'][-1] / cotacao_ibov['Adj Close'][0] -1

print('Carteira: {:.2%}'.format(retorno_carteira))
print('=-'*8)
print('IBOV: {:.2%}'.format(retorno_ibov))
```

## Correlaçao da Carteira com o IBOV

```python
correlacao = valor_investido['Total'].corr(cotacao_ibov['Adj Close'])
print('Correlação: {:.2%} '.format(correlacao * 1))
```
